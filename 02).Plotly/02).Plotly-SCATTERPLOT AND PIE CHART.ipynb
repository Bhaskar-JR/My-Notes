# ScatterPlot & Pie Chart

```python
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import numpy as np
```

```python
#Jupyter setup
init_notebook_mode(connected=True)
```

## ScatterPlot

```python
N = 100
example_x = np.random.rand(N)
example_y = np.random.rand(N)
```

```python
#trace is the name given to the collection of data along with the specification of data in plotly
trace = go.Scatter(
    x = example_x,
    y = example_y,
    mode = 'markers')
```

```python
data = [trace]
iplot(data)
```

## Pie Chart

```python
#Expenses

#breakdown each category 
groups = ['Rent','Food','Bills','Miscellaneous']
#create amount
amount = [1000,518,331,277]
#style
colors = ['#d32c58', '#f9b1ee', '#b7f9b1', '#b1f5f9']

trace = go.Pie(labels=groups, values=amount,
               hoverinfo='label+percent', textinfo='value', 
               textfont=dict(size=25),
               marker=dict(colors=colors, 
                           line=dict(color='#000000', width=3)))

#plot - 
iplot([trace])
```
